# senior citizens - classification
# gender - one hot encoding
# partner - label encoding
# dependents - label encoding
# Phoneservice - label encoding
# Multiplelines - remove no phone service value and add No instead / label encoding    *
# InternetService - one hot encoding
# Online security - remove no internet service value and add No instead / label encoding *
# Online backup - remove no internet service value and add No instead / label encoding *
# Device protection - remove no internet service value and add No instead / label encoding * 
# Tech support - remove no internet service value and add No instead / label encoding     *
# Streaming TV - remove no internet service value and add No instead / label encoding      *
# Streaming Movies - remove no internet service value and add No instead / label encoding    *
# contract - one hot encoding
# paperless billing - label encoding
# payment method - one hot encoding
# churn - (target variable) - label encoding

```python
df['MultipleLines'] = df['MultipleLines'].replace('No phone service','No')
```

```python
df['OnlineSecurity'] = df['OnlineSecurity'].replace('No internet service','No')
```

```python
df['OnlineBackup'] = df['OnlineBackup'].replace('No internet service','No')
```

```python
df['DeviceProtection'] = df['DeviceProtection'].replace('No internet service','No')
```

```python
df['TechSupport'] = df['TechSupport'].replace('No internet service','No')
```

```python
df['StreamingTV'] = df['StreamingTV'].replace('No internet service','No')
```

```python
df['StreamingMovies'] = df['StreamingMovies'].replace('No internet service','No')
```

```python
df['TotalCharges'] = df['TotalCharges'].replace(' ',np.nan)
```

```python
df.info()
```

```python
df['TotalCharges'].unique()
```

```python
df['TotalCharges'] = df['TotalCharges'].astype(float)
```

```python
df['Churn'] = df['Churn'].map({'Yes' : 1, 'No' : 0})
```

```python
df.dropna(inplace = True)
```

```python
df = df.drop(df.sample(n=2100, random_state=42).index)
df['SeniorCitizen'] = df['SeniorCitizen'].map({0:'No', 1:'Yes'})
```

```python
df.info()
```

```python
df.to_csv('cleaned_data.csv', index = False)
```
